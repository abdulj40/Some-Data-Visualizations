```python
# Rows containing duplicate data
duplicate_rows_df = read_data[read_data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
print("0 means no duplicated rows")
```

```python
sl_at_home = [country for country in read_data['home_team'] if country == 'Sierra Leone']
print("Sierra Leone played", len(sl_at_home), "games at home from November, 1872 to July, 2019")
```

```python
sl_away = [country for country in read_data['away_team'] if country == 'Sierra Leone']
print("Sierra Leone played ", len(sl_away), " games away from November, 1872 to July, 2019")
```

```python
# home matches of Sierra Leone and results
sl_home_matches = read_data[read_data['home_team'] == "Sierra Leone"]
sl_home_matches.head()
```

```python
sl_home_matches.tail()
```

```python
# Home wins for Sierra Leone 1872 to 2019
sl_home_wins = read_data[(read_data['home_team'] == "Sierra Leone") & (read_data["home_score"] > read_data["away_score"])]
sl_home_wins.head()
```

```python
# Home losses for Sierra Leone
sl_home_loss = read_data[(read_data['home_team'] == "Sierra Leone") & (read_data["home_score"] < read_data["away_score"])]
sl_home_loss.tail()
```

```python
# Home draws for Sierra Leone
sl_home_draws = read_data[(read_data['home_team'] == "Sierra Leone") & (read_data["home_score"] == read_data["away_score"])]
sl_home_draws.tail()
```

```python
# away matches of Sierra Leone and results
sl_away_matches = read_data[read_data['away_team'] == "Sierra Leone"]
sl_away_matches.head()
```

```python
sl_away_matches.tail()
```

```python
# Away wins for Sierra Leone 1872 to 2019
sl_away_wins = read_data[(read_data['away_team'] == "Sierra Leone") & (read_data["home_score"] < read_data["away_score"])]
sl_away_wins.head()
```

```python
# Away losses for Sierra Leone 1872 to 2019
sl_away_losses = read_data[(read_data['away_team'] == "Sierra Leone") & (read_data["home_score"] > read_data["away_score"])]
sl_away_losses.tail()
```

```python
# Away draws for Sierra Leone 1872 to 2019
sl_away_draws = read_data[(read_data['away_team'] == "Sierra Leone") & (read_data["home_score"] == read_data["away_score"])]
sl_away_draws.tail()
```

```python
import pygal
from pygal.style import LightColorizedStyle as LCS, LightenStyle as LS

my_style = LS('#333366', base_style=LCS)
chart = pygal.Bar(style=my_style, x_label_rotation=45, show_legend=False, width=1100)

chart.title = "Football - Sierra Leone's International Matches Results 1972 - 2019"
chart.x_labels = ['Home Wins', 'Home Losses', 'Home Draws', 'Away Wins', 'Away Losses', 'Away Draws']

plot_dicts = [{'value': sl_home_wins.home_score.count(), 'label': 'Home wins'},
              {'value': sl_home_loss.home_score.count(), 'label': 'Home losses'},
              {'value': sl_home_draws.home_score.count(), 'label': 'Home Draws'},
              {'value': sl_away_wins.home_score.count(), 'label': 'Away wins'},
              {'value': sl_away_losses.home_score.count(), 'label': 'Away losses'},
              {'value': sl_away_draws.home_score.count(), 'label': 'Away draws'}
             ]

chart.add('', plot_dicts)
chart.render_in_browser() # view in new tab in browser
```

```python
# Plotting a Histogram for home matches
read_data.home_team.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of home matches by national team from November, 1872 to July, 2019")
plt.ylabel('Number of matches')
plt.xlabel('Team')
```

```python
# Plotting a Histogram for away matches
read_data.away_team.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of away matches by national team from November, 1872 to July, 2019")
plt.ylabel('Number of matches')
plt.xlabel('Team')
```
